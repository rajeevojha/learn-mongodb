```python
import matplotlib.pyplot as plt

# Sales by Category
sales_by_category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
sales_by_category.plot(kind='bar', title='Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales ($)')
plt.show()
```

```python
# Total sales by Category
sales_by_category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print("\nTotal Sales by Category:\n", sales_by_category)

# Total sales by Region
sales_by_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print("\nTotal Sales by Region:\n", sales_by_region)

# Average sales per order
avg_sales_per_order = df.groupby('Order ID')['Sales'].sum().mean()
print(f"\nAverage Sales per Order: ${avg_sales_per_order:.2f}")
```

```python
df.groupby(['Category', 'Region'])['Sales'].sum().unstack().plot(kind='bar', stacked=True)
plt.title('Sales by Category and Region')
plt.ylabel('Sales ($)')
plt.show()
```

```python
# Filter for Vermont orders
vt_orders = df[df['State'] == 'Vermont']

# Sort by Sales, descending, and get top 5
top_vt_orders = vt_orders.sort_values('Sales', ascending=False).head(5)

# Display key columns
print(top_vt_orders[['Order ID', 'Customer Name', 'Product Name', 'Sales']])
```

```python
# Group by Ship Mode, aggregate Sales (mean) and Order ID (count)
ship_mode_stats = df.groupby('Ship Mode').agg({'Sales': 'mean', 'Order ID': 'nunique'})

# Rename columns for clarity
ship_mode_stats.columns = ['Avg Sales', 'Unique Orders']

# Round Avg Sales for readability
ship_mode_stats['Avg Sales'] = ship_mode_stats['Avg Sales'].round(2)

print(ship_mode_stats)
```

```python
# Pivot table: Sales summed by Category and Region
pivot_sales = pd.pivot_table(df, values='Sales', index='Category', columns='Region', aggfunc='sum', fill_value=0)

# Add a total sales column
pivot_sales['Total'] = pivot_sales.sum(axis=1)

print(pivot_sales)
```

```python
import matplotlib.pyplot as plt
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
# Set Order Date as index and resample by month
monthly_sales = df.set_index('Order Date').resample('ME')['Sales'].sum()

# Plot
monthly_sales.plot(title='Monthly Sales Trend', figsize=(10, 6))
plt.xlabel('Date')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.show()
```

```python
# Define a function to compute shipping delay
def shipping_delay(row):
    return (row['Ship Date'] - row['Order Date']).days

# Apply it to create a new column
df['Shipping Delay'] = df.apply(shipping_delay, axis=1)

# Find top 5 longest delays
longest_delays = df.sort_values('Shipping Delay', ascending=False).head(5)

print(longest_delays[['Order ID', 'Order Date', 'Ship Date', 'Shipping Delay']])
```
